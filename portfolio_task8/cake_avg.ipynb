```python
cake_items = {
    210201: "Flour",
    211709: "Milk",
    211301: "Butter",
    212101: "Sugar",
    211602: "Eggs"
}
```

```python
items_df[items_df['item_id'].isin(cake_items.keys())]
```

```python
full_df = pd.merge(prices_df, items_df[['item_id', 'description']], on='item_id')
```

```python
full_df['quote_date'] = pd.to_datetime(full_df['quote_date'], format='%Y%m')
```

```python
cake_df = full_df[full_df['item_id'].isin(cake_items.keys())]

cake_df
```

```python
cake_avg = cake_df.groupby(['quote_date', 'description', 'item_id']).agg({'price': 'mean'}).reset_index()

cake_avg
```

```python
complete_dates = cake_avg.groupby('quote_date').filter(lambda x: set(cake_items.keys()).issubset(set(x['item_id'])))

earliest_complete_date = complete_dates['quote_date'].min()

last_complete_date = complete_dates['quote_date'].max()
```

```python
cake_avg = cake_avg[(cake_avg['quote_date'] >= earliest_complete_date) & (cake_avg['quote_date'] <= last_complete_date)]
```

```python
cake_avg['item_name'] = cake_avg['item_id'].map(cake_items)

cake_avg
```

```python
cake_avg.to_csv('cake_avg.csv', index=False)
```
