```python
items_df = pd.read_parquet('https://autocpi-public.s3.eu-west-2.amazonaws.com/lrpd/db_item.parquet')
items_df.head()
```

```python
items_df[items_df['description'].str.contains('tampon', case=False)]
```

```python
region_df = pd.read_csv("https://eco-prices-scrapes.s3.eu-west-2.amazonaws.com/teaching/davies_price_data/db_region.csv")

region_df
```

```python
region_codes = {
    2.0: "London",
    3.0: "South East",
    4.0: "South West",
    5.0: "East Anglia",
    6.0: "East Midlands",
    7.0: "West Midlands",
    8.0: "Yorkshire & Humber",
    9.0: "North West",
    10.0: "North",
    11.0: "Wales",
    12.0: "Scotland",
    13.0: "Northern Ireland"
}
```

```python
region_codes.values()
```

```python
full_df = pd.merge(prices_df, items_df[['item_id', 'description']], on='item_id')
```

```python
full_df['region_name'] = full_df['region'].map(region_codes)

full_df
```

```python
tampon_df = full_df[full_df['item_id'] == 520206]

tampon_df.head()
```

```python
tampon_df['region_name'] = tampon_df['region'].map(region_codes)
tampon_df['quote_date'] = pd.to_datetime(tampon_df['quote_date'], format='%Y%m')
tampon_df
```

```python
tampon_prices = prices_df.query("item_id == 520206")

avg_tampon_prices = (
    tampon_df
    .groupby(['quote_date', 'region_name'])
    .agg({'price': ['mean']})
    .reset_index()
)

avg_tampon_prices.columns = ['Date', 'Region', 'Mean']

avg_tampon_prices
```

```python
avg_tampon_prices.to_csv('avg_tampon_prices.csv', index=False)
```
